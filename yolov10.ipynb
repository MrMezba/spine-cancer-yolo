```python
# Dataset à¦à¦° yaml à¦«à¦¾à¦‡à¦²à§‡à¦° à¦ªà¦¾à¦¥ à¦¸à§‡à¦Ÿ à¦•à¦°à§‹
train_yaml = '/content/drive/MyDrive/Final_Defence/data/dataset.yaml'
```

```python
# YOLOv10 à¦®à¦¡à§‡à¦² à¦²à§‹à¦¡ à¦•à¦°à§‹
model = YOLO('yolov10m.pt')
```

```python
# à¦®à¦¡à§‡à¦² à¦Ÿà§à¦°à§‡à¦¨ à¦•à¦°à¦¾à¦“
results = model.train(
    data=train_yaml,    # à¦¡à§‡à¦Ÿà¦¾à¦¸à§‡à¦Ÿ yaml à¦«à¦¾à¦‡à¦² à¦¯à§‡à¦–à¦¾à¦¨à§‡ à¦†à¦›à§‡
    epochs=100,          # à¦®à§‹à¦Ÿ epoch à¦¸à¦‚à¦–à§à¦¯à¦¾
    imgsz=640,          # à¦‡à¦®à§‡à¦œ à¦¸à¦¾à¦‡à¦œ (640x640)
    batch=16,           # à¦ªà§à¦°à¦¤à¦¿ batch-à¦ à¦•à¦¤ à¦›à¦¬à¦¿ à¦¯à¦¾à¦¬à§‡
    workers=4,          # à¦¡à§‡à¦Ÿà¦¾ à¦²à§‹à¦¡ à¦•à¦°à¦¾à¦° à¦œà¦¨à§à¦¯ parallel worker à¦¸à¦‚à¦–à§à¦¯à¦¾
    device=0,           # GPU à¦¡à¦¿à¦­à¦¾à¦‡à¦¸ 0, GPU à¦¨à¦¾ à¦¥à¦¾à¦•à¦²à§‡ 'cpu' à¦¦à¦¿à¦¬à§‡
    project=base_dir,   # à¦¯à§‡à¦–à¦¾à¦¨à§‡ à¦Ÿà§à¦°à§‡à¦¨à¦¿à¦‚ à¦°à¦¾à¦¨ à¦¸à§‡à¦­ à¦¹à¦¬à§‡
    name='train_results' # à¦°à¦¾à¦¨ à¦à¦° à¦¨à¦¾à¦®
)
```

```python
from ultralytics import YOLO
import os

# -----------------------------
# Paths
# -----------------------------
MODEL_PATH = '/content/drive/MyDrive/Final_Defence/yolov10/train_results/weights/best.pt'
DATA_YAML  = '/content/drive/MyDrive/Final_Defence/data/dataset.yaml'
RESULT_DIR = '/content/drive/MyDrive/Final_Defence/yolov10'

os.makedirs(RESULT_DIR, exist_ok=True)

# -----------------------------
# Load trained YOLO model
# -----------------------------
model = YOLO(MODEL_PATH)

# -----------------------------
# Test set evaluation with custom save path
# -----------------------------
results = model.val(
    data=DATA_YAML,
    split='test',
    save=True,        # save prediction images
    save_txt=True,    # save predicted bounding boxes
    save_conf=True,   # save confidence scores
    project=RESULT_DIR,  # custom folder
    name='test_results'  # subfolder inside RESULT_DIR
)
```
