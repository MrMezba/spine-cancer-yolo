```python
# প্রয়োজনীয় লাইব্রেরি ইম্পোর্ট করো
from ultralytics import YOLO
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
base_dir = '/content/drive/MyDrive/Final_Defence/yolov10'
os.makedirs(base_dir, exist_ok=True)
```

```python
# Dataset এর yaml ফাইলের পাথ সেট করো
train_yaml = '/content/drive/MyDrive/Final_Defence/data/dataset.yaml'
```

```python
# YOLOv10 মডেল লোড করো
model = YOLO('yolov10m.pt')
```

```python
# মডেল ট্রেন করাও
results = model.train(
    data=train_yaml,    # ডেটাসেট yaml ফাইল যেখানে আছে
    epochs=100,          # মোট epoch সংখ্যা
    imgsz=640,          # ইমেজ সাইজ (640x640)
    batch=16,           # প্রতি batch-এ কত ছবি যাবে
    workers=4,          # ডেটা লোড করার জন্য parallel worker সংখ্যা
    device=0,           # GPU ডিভাইস 0, GPU না থাকলে 'cpu' দিবে
    project=base_dir,   # যেখানে ট্রেনিং রান সেভ হবে
    name='train_results' # রান এর নাম
)
```

```python
# Training results CSV ফাইল থেকে শেষ ১০টি রেকর্ড পড়ো ও প্রিন্ট করো
csv_path = os.path.join(base_dir, 'train_results', 'results.csv')
df = pd.read_csv(csv_path)
print(df.tail(10))
```

```python
from ultralytics import YOLO
import os

# -----------------------------
# Paths
# -----------------------------
MODEL_PATH = '/content/drive/MyDrive/Final_Defence/yolov10/train_results/weights/best.pt'
DATA_YAML  = '/content/drive/MyDrive/Final_Defence/data/dataset.yaml'
RESULT_DIR = '/content/drive/MyDrive/Final_Defence/yolov10'

os.makedirs(RESULT_DIR, exist_ok=True)

# -----------------------------
# Load trained YOLO model
# -----------------------------
model = YOLO(MODEL_PATH)

# -----------------------------
# Test set evaluation with custom save path
# -----------------------------
results = model.val(
    data=DATA_YAML,
    split='test',
    save=True,        # save prediction images
    save_txt=True,    # save predicted bounding boxes
    save_conf=True,   # save confidence scores
    project=RESULT_DIR,  # custom folder
    name='test_results'  # subfolder inside RESULT_DIR
)
```
