```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# ---- Training Box Loss ----
for model_name, file_name in models.items():
    file_path = os.path.join(csv_dir, file_name)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        df.columns = df.columns.str.strip()
        if "train/box_loss" in df.columns:
            sns.lineplot(ax=axes[0], x=df["epoch"], y=df["train/box_loss"], label=model_name)

axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Loss")
axes[0].grid(False)
axes[0].legend()
axes[0].set_yticks(np.arange(0, axes[0].get_ylim()[1] + 0.5, 0.5))

# ---- (a) ঠিক Epoch এর নিচে ----
axes[0].set_xlabel("Epoch\n\n(a)", fontsize=12)

# ---- Validation Box Loss ----
for model_name, file_name in models.items():
    file_path = os.path.join(csv_dir, file_name)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        df.columns = df.columns.str.strip()
        if "val/box_loss" in df.columns:
            sns.lineplot(ax=axes[1], x=df["epoch"], y=df["val/box_loss"], label=model_name)


axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Loss")
axes[1].grid(False)
axes[1].legend()
axes[1].set_yticks(np.arange(0, axes[1].get_ylim()[1] + 0.5, 0.5))

# ---- (b) ঠিক Epoch এর নিচে ----
axes[1].set_xlabel("Epoch\n\n(b)", fontsize=12)

# Extra space
plt.subplots_adjust(wspace=0.35)

plt.show()
```

```python
import os
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

for model_name, file_name in models.items():
    file_path = os.path.join(csv_dir, file_name)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        df.columns = df.columns.str.strip()

        # ---- mAP50 ----
        if "metrics/mAP_0.5" in df.columns:
            sns.lineplot(ax=axes[0], x=df["epoch"], y=df["metrics/mAP_0.5"], label=model_name)
        elif "metrics/mAP50(B)" in df.columns:
            sns.lineplot(ax=axes[0], x=df["epoch"], y=df["metrics/mAP50(B)"], label=model_name)

        # ---- mAP50-95 ----
        if "metrics/mAP_0.5:0.95" in df.columns:
            sns.lineplot(ax=axes[1], x=df["epoch"], y=df["metrics/mAP_0.5:0.95"], label=model_name)
        elif "metrics/mAP50-95(B)" in df.columns:
            sns.lineplot(ax=axes[1], x=df["epoch"], y=df["metrics/mAP50-95(B)"], label=model_name)

# ---- Titles & Labels ----
axes[0].set_title("mAP50 Comparison", fontsize=14)
axes[0].set_xlabel("Epoch\n\n(a)", fontsize=13)
axes[0].set_ylabel("Accuracy", fontsize=13)
axes[0].grid(False)
axes[0].legend()
axes[0].set_yticks(np.arange(0, 1.05, 0.2))

axes[1].set_title("mAP50-95 Comparison", fontsize=14)
axes[1].set_xlabel("Epoch\n\n(b)", fontsize=13)
axes[1].set_ylabel("Accuracy", fontsize=13)
axes[1].grid(False)
axes[1].legend()
axes[1].set_yticks(np.arange(0, 0.65, 0.1))

plt.subplots_adjust(wspace=0.35)
plt.show()
```

```python
import os
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

def find_column(columns, keywords):
    for col in columns:
        for kw in keywords:
            if kw.lower() in col.lower():
                return col
    return None

# ---- Precision ----
for model_name, file_name in models.items():
    file_path = os.path.join(csv_dir, file_name)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        df.columns = df.columns.str.strip()
        precision_col = find_column(df.columns, ["precision", "metrics/precision"])
        if precision_col:
            sns.lineplot(ax=axes[0], x=df["epoch"], y=df[precision_col], label=model_name)

axes[0].set_title("Precision", fontsize=14)
axes[0].set_xlabel("Epoch\n\n(c)", fontsize=13)  # নিচে (a)
axes[0].set_ylabel("Precision", fontsize=13)
axes[0].grid(False)
axes[0].legend()
axes[0].set_yticks(np.arange(0, 1.05, 0.1))

# ---- Recall ----
for model_name, file_name in models.items():
    file_path = os.path.join(csv_dir, file_name)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        df.columns = df.columns.str.strip()
        recall_col = find_column(df.columns, ["recall", "metrics/recall"])
        if recall_col:
            sns.lineplot(ax=axes[1], x=df["epoch"], y=df[recall_col], label=model_name)

axes[1].set_title("Recall", fontsize=14)
axes[1].set_xlabel("Epoch\n\n(d)", fontsize=13)  # নিচে (b)
axes[1].set_ylabel("Recall", fontsize=13)
axes[1].grid(False)
axes[1].legend()
axes[1].set_yticks(np.arange(0, 0.9, 0.1))

# ---- Adjust spacing ----
plt.subplots_adjust(wspace=0.35)
plt.show()
```
